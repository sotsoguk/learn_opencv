```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotCV(image):
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.xticks([])
    plt.yticks([])
```

```python
image = cv2.imread('data/img1.jpg')
#cv2.imshow("test",image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
#plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plotCV(image)
```
